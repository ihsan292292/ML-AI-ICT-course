```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn import svm
from sklearn.metrics import accuracy_score
```

Data Collection & Processing

```python
# load the data from csv file to Pandas DataFrame
parkinson_data = pd.read_csv('/home/ihsan/Documents/GitHub/ML-AI-ICT-course/ML Projects/Diabetes prediction/Parkinson Disease Detection/parkinsons.data')
```

Data Preprocessing

Seperating the features and the targets

```python
x = parkinson_data.drop(columns=['name','status'], axis=0)
y = parkinson_data['status']
```

Spliting the data

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
print(x.shape, x_train.shape, x_test.shape)
```

Data Standardization

```python
scaler = StandardScaler()
```

```python
scaler.fit(x_train) # fit x_train alone
```

```python
x_train = scaler.transform(x_train)
x_test = scaler.transform(x_test)
```

```python
print(x_test)
```

Model Training

Support Vector Machine


```python
# support vectors are the datapoints that are close to the hyperplane
model = svm.SVC(kernel='linear') # Support vector classifier
```

```python
# training the svm model with training data

model.fit(x_train, y_train)
```

Modal Evaluation

Accuracy score

```python
# accuracy score in training data 
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(y_train, x_train_prediction)
```

```python
print("accuracy_score of training data: ",training_data_accuracy)
```

```python
x_test_prediction = model.predict(x_test)
testing_data_accuracy = accuracy_score(y_test, x_test_prediction)
```

```python
print("accuracy_score of testing data: ",testing_data_accuracy)
```

Building Predictive System

```python
input_data = (202.26600,211.60400,197.07900,0.00180,0.000009,0.00093,0.00107,0.00278,0.00954,0.08500,0.00469,0.00606,0.00719,0.01407,0.00072,32.68400,0.368535,0.742133,-7.695734,0.178540,1.544609,0.056141)
# changing input data to a numpy array 
input_data_as_numpy_array = np.asarray(input_data)

# reshape the numpy array
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1) # value for one data point alone

std_data = scaler.transform(input_data_reshaped)

prediction = model.predict(std_data)
print(prediction)

if (prediction[0] == 0):
    print("The person does not have parkinsons disease")
else:
    print("The person have Parkinson")
```
