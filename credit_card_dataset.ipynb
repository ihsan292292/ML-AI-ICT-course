```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv(r"C:\GitHub\ICT - ML&AI\Dataset\credit_card_default.csv")
data
```

```python
data.columns
```

```python
# change titles to lowercase
data = data.rename(columns= lambda x:x.lower())
```

```python
# Rename Column
data = data.rename(columns={"default payment next month":"default"})
data
```

```python
data['education'].value_counts() # Check how many values in education to catagorize
```

```python
data["male"] = (data['sex'] == 1).astype(int) # If sex=1 create new column with binary values 0 and 1
data['grad_school'] = (data['education'] == 1).astype(int) # If education=1 then create a new column with binary values 0 and 1
data['university'] = (data['education'] == 2).astype(int) # If education=1 then create a new column with binary values 0 and 1
data['married'] = (data['marriage'] == 1).astype(int) # If marriage=1 then create a new column with binary values 0 and 1
```

```python
data
```

```python
data[['male','grad_school','university','married']]
```

```python
data.columns
```

```python
'''bill_amt_features = ['bill_amt1', 'bill_amt2',
       'bill_amt3', 'bill_amt4', 'bill_amt5', 'bill_amt6']'''

bill_amt_features = ['bill_amt' + str(i) for i in range(1,7)] # add all bill amount to one column
```

```python
pay_amt_features = ["pay_amt"+ str(i) for i in range(1,7)] # add all pay amount to one column
binary_features = ['male','grad_school','university','married'] # add binary features to one column
```

```python
pay_features = ["pay_"+ str(i) for i in range(1,7)] # add all pay features to one column
```

```python
num_features = ["limit_bal","age"] +bill_amt_features + pay_amt_features + pay_features # add all numerical features 
```

```python
x = data[num_features + binary_features] 
```

```python
y = data['default']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size= 0.2,random_state= 134)
```

```python
sc_features = ["limit_bal","age"] +bill_amt_features + pay_amt_features 
```

```python
from sklearn.preprocessing import StandardScaler # preprocessing for numerical features
```

```python
sc = StandardScaler()
```

```python
x_train.loc[:,sc_features]
```

```python
x_train.loc[:,sc_features] = sc.fit_transform(x_train[sc_features])
```

```python
x_test.loc[:,sc_features] = sc.fit_transform(x_test[sc_features])
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score,r2_score
```

KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
from sklearn.model_selection import KFold,cross_val_score
```

```python
x
```

```python
knn_model=KNeighborsClassifier(n_neighbors=9)
```

```python
k_folds = KFold(n_splits = 5)
scores = cross_val_score(knn_model, x, y, cv = k_folds)
```

```python
print("Cross Validation Scores: ", scores)
print("Average CV Score: ", scores.mean())
print("Number of CV Scores used in Average: ", len(scores))
```

```python
knn_model.fit(x_train,y_train)
```

```python
knn_pred=knn_model.predict(x_test)
```

```python
print('Accuracy score is',accuracy_score(y_test,knn_pred)*100)
```

SVM

```python
clf = svm.SVC(kernel = 'linear') # Create a svm classifier
```

```python
clf.fit(x_train,y_train)
```

```python
y_pred = clf.predict(x_test)
```

Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.datasets import make_regression
```

```python
x, y = make_regression(n_samples=100,
                       n_features=10,
                       n_informative=5,
                       n_targets=1,
                       random_state=42)
```

```python
# Instantiate Gradient Boosting Regressor
gbr = GradientBoostingRegressor(n_estimators = 200, max_depth = 1, random_state = 1)
```

```python
gbr.fit(x_train,y_train)
```

```python
gbr_prd = gbr.predict(x_test)
```

```python
print("Gradient Boosting - R2: ",
      r2_score(y_test, gbr_prd))
```

XG boosting

```python
from xgboost.XGBRegressor import XGBRegressor
xgr = XGBRegressor()
xgr.fit(x_train, y_train)
y_pred2 = xgr.predict(x_test)
print("XGBoost - R2: ",
      r2_score(y_test, y_pred2))
```

ADA Boosting

```python

from sklearn.ensemble import AdaBoostRegressor
adr = AdaBoostRegressor()
adr.fit(x_train, y_train)
y_pred3 = adr.predict(x_test)
print("AdaBoost - R2: ",
      r2_score(y_test, y_pred3))
```

Hyperparameter tuning by gridsearch and randomserach

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
```

```python
param_grid_knn = {
    'n_neighbors': np.arange(5,15),
    'p': (1,2),
    'weights': ('uniform', 'distance'),
    'metric': ('minkowski', 'chebyshev')
} 
knn_gscv = KNeighborsClassifier()
knn_gscv = GridSearchCV(knn_gscv, param_grid=param_grid_knn, scoring='accuracy', cv=5)
knn_gscv.fit(x_train, y_train)
```

```python
print("Best hyperparameters:", knn_gscv.best_params_)
print("Best accuracy score:", knn_gscv.best_score_)
```
