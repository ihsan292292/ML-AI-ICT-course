```python
# Import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
nc_data = pd.read_csv(r"C:\GitHub\ICT - ML&AI\Dataset\NCI60_X.csv",index_col=0)
nc_data
```

```python
# Standardize the data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(nc_data)
```

```python
# Import Kmeans
from sklearn.cluster import KMeans
```

```python
KMeans_inst = KMeans(n_clusters=6)  # Create Kmeans instance
```

```python
KMeans_inst.fit(x) #Aplly Kmeans model
```

```python
# import Agglomerativemodel
from sklearn.cluster import AgglomerativeClustering
```

```python
# Apply Agglomerative clustering
agglo = AgglomerativeClustering(n_clusters=3)
agglo.fit(nc_data)
```

```python
from sklearn.metrics import silhouette_score
```

```python
# Evaluate KMeans and agglomerative clustering performance
kmc_score = silhouette_score(x, KMeans_inst.labels_, metric = 'euclidean')
agg_score = silhouette_score(x, agglo.labels_, metric = 'euclidean')
```

```python
from sklearn.decomposition import PCA
```

```python
pc = PCA(n_components=25) #Apply PCA and reduce the features to 25
```

```python
x_pca = pc.fit_transform(x)
```

```python
x_pca.shape
```

```python
pd.DataFrame(x_pca)
```

```python
nc_data_pca = pd.DataFrame(x_pca)
```

```python
# Apply KMeans clustering on PCA data
kmeans_nc_pca = KMeans(n_clusters=3, random_state=42)
kmeans_nc_pca.fit(nc_data_pca)
```

```python
# Apply Agglomerative clustering on PCA data
agglo_nc_pca = AgglomerativeClustering(n_clusters=3)
agglo_nc_pca.fit(nc_data_pca)
```

```python
# Evaluate KMeans and agglomerative clustering performance after pca
kmc_score_pca = silhouette_score(x_pca, kmeans_nc_pca.labels_, metric = 'euclidean')
agg_score_pca = silhouette_score(x_pca, agglo_nc_pca.labels_, metric = 'euclidean')
```

Comparison

```python
comparison = (('Raw Data', kmc_score, agg_score),
      ('PCA reduced', kmc_score_pca, agg_score_pca))
```

```python
comparison_df = pd.DataFrame(comparison, columns = ["Dataset", "KMeans", "Agglomerative"])
comparison_df.set_index('Dataset', inplace=True)
comparison_df
```

```python
y_kmc = KMeans_inst.fit_predict(x)
plt.scatter(x[y_kmc==0, 0], x[y_kmc==0, 1], s=50, c='red', label ='KMC Cluster 1')
plt.scatter(x[y_kmc==1, 0], x[y_kmc==1, 1], s=50, c='blue', label ='KMC Cluster 2')
plt.scatter(x[y_kmc==2, 0], x[y_kmc==2, 1], s=50, c='green', label ='KMC Cluster 3')
plt.scatter(x[y_kmc==3, 0], x[y_kmc==3, 1], s=50, c='cyan', label ='KMC Cluster 4')
#Plot the centroid. This time we're going to use the cluster centres  #attribute that returns here the coordinates of the centroid.
plt.scatter(KMeans_inst.cluster_centers_[:, 0], KMeans_inst.cluster_centers_[:, 1], s=100, c='yellow',edgecolors='k', label = 'Centroids KMC')
plt.title('KMeans Clusters')
plt.show()
```
